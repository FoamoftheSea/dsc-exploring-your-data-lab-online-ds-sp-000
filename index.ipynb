## Check for Linearity

Recall that one assumption in linear regression is that the target variable is linearly related to the input features. As shown in the previous lesson, you can use the `sns.jointplot()` function to investigate whether this relation holds true for the various predictors on hand.

```python
sns.jointplot(df.num_reviews, df.list_price, kind='reg');
```

```python
sns.jointplot(df.piece_count, df.list_price, kind='reg');
```

```python
sns.jointplot(df.play_star_rating, df.list_price, kind='reg');
```

```python
sns.jointplot(df.star_rating, df.list_price, kind='reg');
```

```python
sns.jointplot(df.val_star_rating, df.list_price, kind='reg');
```

## Checking for Multicollinearity

It's also important to make note of whether your predictive features will result in multicollinearity in the resulting model. While definitive checks for multicollinearity require analyzing the resulting model, predictors with overly high pairwise-correlation (r > .65) are almost certain to produce multicollinearity in a model. With that, take a minute to generate the pairwise (pearson) correlation coefficients of your predictive features and visualize these coefficients as a heatmap.

```python
from IPython.display import display
```

```python
corr = df.select_dtypes('float64').corr()
display(corr)
sns.heatmap(corr)
plt.show()
display(corr>.65)
```

```python
# We can see that there is a strong correlation between star_rating and val_star_rating, so this
# could be dealt with either by dropping one or averaging them together, but from the
# visualizations above, they promise to be weak predictors, so they may both get dropped.
# The play_star_rating also did not show a linear correlation
# list_price and num_reviews are correlated, but not in excess of 0.65 r score
```
